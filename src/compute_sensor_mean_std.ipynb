# Compute sensor mean and standard deviation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

PATH = '/Users/chpfei/LocalData/Datascience/FlyingCar/04_estimation_TODO/project/FCND-Estimation-CPP/config/log/'

df = pd.read_csv(PATH+'Graph1.txt')

x = df['time']
y = df[' Quad.GPS.X']

plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.title('GPS mean={:.4f}, std:{:.4f}, abs(mean)+std: {:.4f}'.format(np.mean(y), 
                                                                     np.std(y), 
                                                                     np.std(y)+np.abs(np.mean(y))))


df = pd.read_csv(PATH+'Graph2.txt')

x = df['time']
y = df[' Quad.IMU.AX']

plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.title('IMU mean={:.4f}, std:{:.4f}, abs(mean)+std: {:.4f}'.format(np.mean(y), 
                                                                     np.std(y), 
                                                                     np.std(y)+np.abs(np.mean(y))))
```
